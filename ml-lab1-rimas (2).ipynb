# Plotting the logsigmoid function
# a) f(x)=1/(1+e^-x)

```python
#here I'm implementing the log sigmoid function, the exp function is for e
def logsigmoid(x):
    return 1/(1+np.exp(-x))

#for x between -5 and 5
x = np.linspace(-5, 5, 100)

# Compute corresponding y values so that when plotting i have x and y
y = logsigmoid(x)

#plotting the function
plt.plot(x,y, color = 'purple')
plt.xlabel('x/input values')
plt.ylabel('y/output values')
plt.title('Plot of f(x) = 1 / (1 + e^(-x))')
plt.grid(True)
plt.show()
```

# Plotting the log function
# b) f(x)=log(1+e^x)

```python
#here i'm implementing the log function by np.log
def logarithmic(x):
    return np.log(1+np.exp(-x))

#for x between -5 and 5
x=np.linspace(-5,5,100)

#assigning y
y=logarithmic(x)


#plotting the function
plt.plot(x,y, color='purple')
plt.xlabel('x/ input of logarithmic function ')
plt.ylabel('y/output of logarithmic function')
plt.title('plot of log function f(x)=log(1+e^x)')
plt.grid(True)
plt.show()
```

# c) 

```python

# Generate x values
x = np.linspace(-5, 5, 100)

fig, ax = plt.subplots(1, 2)

# Plot f(x) = 1 / (1 + e^(-x)) in the first subplot (ax[0])
ax[0].plot(x, logsigmoid(x), color='purple')
ax[0].set_xlabel('x')
ax[0].set_ylabel('Y')
ax[0].set_title('Plot of logsigmoid = 1 / (1 + e^(-x))')
ax[0].grid(True)

# Plot f(x) = log(1 + e^x) in the second subplot (ax[1])
ax[1].plot(x, logarithmic(x), color='purple')
ax[1].set_xlabel('x')
ax[1].set_ylabel('Y')
ax[1].set_title('Plot of logarithmic(x) = log(1 + e^x)')
ax[1].grid(True)

# Adjust the layout and display the figure
plt.tight_layout()

plt.subplots_adjust(wspace=0.5)

plt.show()
```
